```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from itertools import chain, combinations
```

```python
meta = pd.read_csv("/content/drive/MyDrive/Pewpularuty/train.csv")
```

```python
meta.drop(['Id'],axis =1,inplace= True)
```

```python
metacombi = meta[[col for col in list(meta.columns) if col not in ['Pawpularity','Id']]]
```

```python
metacombi_final = meta[['Pawpularity']]
```

```python
# create combination of all metadata columns exclusively 
# summing up values in those colums for each picture
combs = list(chain.from_iterable(combinations(metacombi.columns, i)
                                 for i in range(2, len(metacombi.columns) + 1)))
for cols in combs:
    metacombi_final['_'.join(cols)] = metacombi.loc[:, cols].sum(axis=1)
```

```python
# resulted data frame with metadata columns combination, i.e 4084 columns 
metacombi_final.head()
```

```python
# performing correlation coefficient caluculation of all metadata columns combination and target score 
# aim is to find combination of metadata that may correlate with the scoe 
corr_pic = metacombi_final.corr(method ='pearson')
corr_pic
```

```python
# saving data
corr_pic.to_csv('corr_pic.csv')
from google.colab import files
files.download("corr_pic.csv")
```
